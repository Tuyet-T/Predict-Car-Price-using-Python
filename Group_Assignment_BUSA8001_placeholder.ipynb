**Task 2, Question 1**: Clean **all** numerical features so that they can be used in training algorithms.
```python
# Remove units and convert data from string to numeric or NA when not convertable
df_train['back_legroom'] = df_train['back_legroom'].str.replace(' in', '')
df_train['back_legroom'] = pd.to_numeric(df_train['back_legroom'], errors = 'coerce')

df_train['front_legroom'] = df_train['front_legroom'].str.replace(' in', '')
df_train['front_legroom'] = pd.to_numeric(df_train['front_legroom'], errors = 'coerce')

df_train['fuel_tank_volume'] = df_train['fuel_tank_volume'].str.replace(' gal', '')
df_train['fuel_tank_volume'] = pd.to_numeric(df_train['fuel_tank_volume'], errors = 'coerce')

df_train['height'] = df_train['height'].str.replace(' in', '')
df_train['height'] = pd.to_numeric(df_train['height'], errors = 'coerce')

df_train['length'] = df_train['length'].str.replace(' in', '')
df_train['length'] = pd.to_numeric(df_train['length'], errors = 'coerce')

df_train['maximum_seating'] = df_train['maximum_seating'].str.replace(' seats', '')
df_train['maximum_seating'] = pd.to_numeric(df_train['maximum_seating'], errors = 'coerce')

df_train['wheelbase'] = df_train['wheelbase'].str.replace(' in', '')
df_train['wheelbase'] = pd.to_numeric(df_train['wheelbase'], errors = 'coerce')

df_train['width'] = df_train['width'].str.replace(' in', '')
df_train['width'] = pd.to_numeric(df_train['width'], errors = 'coerce')

# on test dataset too
df_test['back_legroom'] = df_test['back_legroom'].str.replace(' in', '')
df_test['back_legroom'] = pd.to_numeric(df_test['back_legroom'], errors = 'coerce')

df_test['front_legroom'] = df_test['front_legroom'].str.replace(' in', '')
df_test['front_legroom'] = pd.to_numeric(df_test['front_legroom'], errors = 'coerce')

df_test['fuel_tank_volume'] = df_test['fuel_tank_volume'].str.replace(' gal', '')
df_test['fuel_tank_volume'] = pd.to_numeric(df_test['fuel_tank_volume'], errors = 'coerce')

df_test['height'] = df_test['height'].str.replace(' in', '')
df_test['height'] = pd.to_numeric(df_test['height'], errors = 'coerce')

df_test['length'] = df_test['length'].str.replace(' in', '')
df_test['length'] = pd.to_numeric(df_test['length'], errors = 'coerce')

df_test['maximum_seating'] = df_test['maximum_seating'].str.replace(' seats', '')
df_test['maximum_seating'] = pd.to_numeric(df_test['maximum_seating'], errors = 'coerce')

df_test['wheelbase'] = df_test['wheelbase'].str.replace(' in', '')
df_test['wheelbase'] = pd.to_numeric(df_test['wheelbase'], errors = 'coerce')

df_test['width'] = df_test['width'].str.replace(' in', '')
df_test['width'] = pd.to_numeric(df_test['width'], errors = 'coerce')
```

`We found that the columns to be cleaned here are back_legroom, front_legroom, fuel_tank_volume, height, length, maximum_seating, wheelbase, width. We extract the numbers in the columns by replacing the text values with blank. Then we use 'pd.to_numeric' to convert the type to numeric.`

**Task 2, Question 2** Create at least 5 new features from the existing numerical variables which contain multiple items of information, for example you could extract maximum torque and torque rpm from the torque variable.  
(2 marks)

```python
# Split 'power' and 'torque' into 2, extract number from 'transmission'
df_train[['power_hp','power_rpm']] = df_train.power.apply(lambda x: pd.Series(str(x).split(" hp @ ")))
df_train['power_rpm'] = df_train['power_rpm'].str.replace('RPM', '')
df_train['power_rpm'] = df_train['power_rpm'].str.replace(',', '')
df_train['power_hp'] = pd.to_numeric(df_train['power_hp'], errors = 'coerce')
df_train['power_rpm'] = pd.to_numeric(df_train['power_rpm'], errors = 'coerce')

df_test[['power_hp','power_rpm']] = df_test.power.apply(lambda x: pd.Series(str(x).split(" hp @ ")))
df_test['power_rpm'] = df_test['power_rpm'].str.replace('RPM', '')
df_test['power_rpm'] = df_test['power_rpm'].str.replace(',', '')
df_test['power_hp'] = pd.to_numeric(df_test['power_hp'], errors = 'coerce')
df_test['power_rpm'] = pd.to_numeric(df_test['power_rpm'], errors = 'coerce')


df_train[['torque_lbft','torque_rpm']] = df_train.torque.apply(lambda x: pd.Series(str(x).split(" lb-ft @ ")))
df_train['torque_rpm'] = df_train['torque_rpm'].str.replace('RPM', '')
df_train['torque_rpm'] = df_train['torque_rpm'].str.replace(',', '')
df_train['torque_lbft'] = pd.to_numeric(df_train['torque_lbft'], errors = 'coerce')
df_train['torque_rpm'] = pd.to_numeric(df_train['torque_rpm'], errors = 'coerce')

df_test[['torque_lbft','torque_rpm']] = df_test.torque.apply(lambda x: pd.Series(str(x).split(" lb-ft @ ")))
df_test['torque_rpm'] = df_test['torque_rpm'].str.replace('RPM', '')
df_test['torque_rpm'] = df_test['torque_rpm'].str.replace(',', '')
df_test['torque_lbft'] = pd.to_numeric(df_test['torque_lbft'], errors = 'coerce')
df_test['torque_rpm'] = pd.to_numeric(df_test['torque_rpm'], errors = 'coerce')

# Split transmission_display
df_train['transmission_display'] = df_train['transmission_display'].fillna(df_train['transmission_display'].mode())
df_test['transmission_display'] = df_test['transmission_display'].fillna(df_test['transmission_display'].mode())

df_train[['transmission_speed','transmission_type']] = df_train.transmission_display.apply(lambda x: pd.Series(str(x).split("-Speed ")))
df_train['transmission_speed'] = pd.to_numeric(df_train['transmission_speed'], errors = 'coerce')

df_test[['transmission_speed','transmission_type']] = df_test.transmission_display.apply(lambda x: pd.Series(str(x).split("-Speed ")))
df_test['transmission_speed'] = pd.to_numeric(df_test['transmission_speed'], errors = 'coerce')


# Convert the 'listed_date' column to datetime
df_train['listed_date'] = pd.to_datetime(df_train['listed_date'], format='%d-%m-%y')

# Extract the month and year from the 'listed_date' column
df_train['listed_month'] = df_train['listed_date'].dt.month
df_train['listed_year'] = df_train['listed_date'].dt.year

# Convert the 'listed_date' column to datetime
df_test['listed_date'] = pd.to_datetime(df_test['listed_date'], format='%Y-%m-%d')

# Extract the month and year from the 'listed_date' column
df_test['listed_month'] = df_test['listed_date'].dt.month
df_test['listed_year'] = df_test['listed_date'].dt.year

```

`We found that we can extract new features in the columns: power, torque, transmission display and listed_date. 
For 'power', we split it to power_hp and power _rpm. For torque, we split it to torque_lbft and torque_rpm.
After spliting them we extract the number by removing the unit and convert the type to numeric.`

`For transmission display, first we fill the missing values by mode, then we split it to transmission_speed and transmission_type. The process is same as above.
For the listed_date, first we convert it to datetime format, then we extract month and year into different columns called listed_month and listed_year.`

## Data preprocessing

- Put these variables in `X_train` and `X_test` respectively, and put `price` from `df_train` in `y_train`

```python
# Pick variables to exclude, scale the data
X_train = df_train.drop(bad_index.tolist(), axis=1)
X_train = X_train.drop(['vin', 'price', 'model_name'], axis=1)

X_test = df_test.drop(bad_index.tolist(), axis=1)
X_test = X_test.drop(['vin', 'model_name'], axis=1)

y_train = df_train['price']
```

- Next, we scale both dataset using `StandardScaler`

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train_scaled = sc.transform(X_train)
X_test_scaled = sc.transform(X_test)
```

## XGBoost 

- Before fitting the model, we tune the hyperparameter of each model using `GridSearchCV` with `scoring = neg_mean_squared_error`  and `cv = 5`
- Import `xgb` and create the model. Define the parameter grid for hyperparameter tuning
- Initialize `GridSearchCV` with model, parametric grid, and scoring metric
- Perform the grid search
- Obtain the best set of hyperparameters

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV

# Create model
xgb_model = xgb.XGBRegressor()

# Define parameter grid for hyperparameter tuning
param_grid = {
    'learning_rate': [0.2],
    'gamma': [1, 5, 10],
    'max_depth': [3, 4, 5],
    'min_child_weight': [0.1, 1, 10],
    'n_estimators': [100, 200, 300],
}

# Initialise GridSearchCV
grid_search = GridSearchCV(
    estimator=xgb_model,
    param_grid=param_grid,
    scoring='neg_mean_squared_error',
    cv=5,
    n_jobs=-1,
)

# Perform hyperparameter tuning
grid_search.fit(X_train_scaled, y_train)

# Print the best hyperparameters
print("Best Hyperparameters: ", grid_search.best_params_)
```

- Now we are ready to fit the model using the hyperparameters calculated above: {'gamma': 1, 'learning_rate': 0.2, 'max_depth': 5, 'min_child_weight': 10, 'n_estimators': 300}
- Fit to the training dataset then predict using the test dataset
- Transform to the format required and export to `'submission_xgboost.csv'`

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV

# Use best hyperparameters
best_xgb_model = xgb.XGBRegressor(
    gamma=1,
    learning_rate=0.2,
    max_depth=5,
    min_child_weight=10,
    n_estimators=300,
)

# Fit the model to training data
best_xgb_model.fit(X_train_scaled, y_train)

# Make prediction with test data
y_test = best_xgb_model.predict(X_test_scaled)

# Export to csv file
result = pd.DataFrame(y_test, index = df_test['vin'])
result.rename(columns = {0:'price'}, inplace = True)
print(result.head())

result.to_csv('submission_xgboost.csv')
```
